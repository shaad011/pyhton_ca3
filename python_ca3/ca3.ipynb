### Q10

```python
df['PREMIUM'] = df['SOLD PRICE'] - df['BASE PRICE']
max_premium_player = df.loc[df['PREMIUM'].idxmax()]
print("\nPlayer with the maximum premium:")
print(f"Player Name: {max_premium_player['PLAYER NAME']}")
print(f"Sold Price: {max_premium_player['SOLD PRICE']}")
print(f"Base Price: {max_premium_player['BASE PRICE']}")
print(f"Premium: {max_premium_player['PREMIUM']}")
```

### Q11

```python
max_premium_value = df['PREMIUM'].max()

max_premium_players = df[df['PREMIUM'] == max_premium_value]

print("\nPlayers with the maximum premium:")
for index, player in max_premium_players.iterrows():
    print(f"Player Name: {player['PLAYER NAME']}, Sold Price: {player['SOLD PRICE']}, Base Price: {player['BASE PRICE']}, Premium: {player['PREMIUM']}\n")
```

### Q12

```python
filtered_df = df[df['AGE'].isin([1, 2, 3])]
average_sold_price1 = filtered_df.groupby('AGE')['SOLD PRICE'].mean().reset_index()

print("Average SOLD PRICE for each age (1, 2, 3):")
average_sold_price1
```

### Q13

```python
average_sold_price2 = filtered_df.groupby(['AGE', 'PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
print("Average SOLD PRICE for different playing roles in ages 1, 2, and 3:")
pd.set_option('display.float_format', '{:.2f}'.format)
average_sold_price2
```
